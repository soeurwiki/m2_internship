```python
import pandas as pd
```

```python
markers = pd.read_csv('top_markers_metabo.csv',sep=',',header=0)
markers
```

```python
cl0 = markers['cl0'].unique(); print(len(cl0))
cl1 = markers['cl1'].unique(); print(len(cl1))
cl2 = markers['cl2'].unique(); print(len(cl2))
cl3 = markers['cl3'].unique(); print(len(cl3))
cl4 = markers['cl4'].unique(); print(len(cl4))
cl5 = markers['cl5'].unique(); print(len(cl5))
cl6 = markers['cl6'].unique(); print(len(cl6))
cl7 = markers['cl7'].unique(); print(len(cl7))
cl8 = markers['cl8'].unique(); print(len(cl8))
cl9 = markers['cl9'].unique(); print(len(cl9))
cl10 = markers['cl10'].unique(); print(len(cl10))
```

```python
markers = pd.DataFrame.from_dict({'cl0' : cl0, 'cl1' : cl1, 'cl2' : cl2, 'cl3' : cl3,
                        'cl4' : cl4, 'cl5' : cl5, 'cl6' : cl6, 'cl7' : cl7,
                        'cl8' : cl8, 'cl9' : cl9, 'cl10' : cl10}, orient='index').transpose()
markers
```

```python
clusters_comp = [markers['cl0'],markers['cl1'],markers['cl2'],markers['cl3'],
                    markers['cl4'],markers['cl5'],markers['cl6'],markers['cl7'],
                    markers['cl8'],markers['cl9'],markers['cl10']]
```

```python
# get genes that are present in more than one cluster
result = pd.concat(clusters_comp); print(len(result))
result = result[result.duplicated()].dropna().unique()

len(result)
```

```python
# get all genes present in markers
unique = pd.concat(clusters_comp).unique()
unique = [item for item in unique if not(pd.isnull(item)) == True] ## retirer les valeurs None/Nan
len(unique)
```

```python
# get genes that are seen in only one cluster
main_list = list(set(unique) - set(result))
main_list = [item for item in (main_list) if not(pd.isnull(item)) == True]

len(main_list)
```

```python
cluster_markers = []
for i in range(11):
    cluster = 'cl' + str(i)
    cluster_markers.append(list(set(markers[cluster]) - set(result)))
    cluster_markers[i]= [item for item in cluster_markers[i] if not(pd.isnull(item)) == True]
    print(cluster + " "+ str(len([item for item in markers[cluster] if not(pd.isnull(item)) == True])) +" "+ str(len(cluster_markers[i])))
```

```python
# Define a dictionary

data = {'cluster_0': cluster_markers[0],
        'cluster_1': cluster_markers[1],
        'cluster_2': cluster_markers[2],
        'cluster_3': cluster_markers[3],
        'cluster_4': cluster_markers[4],
        'cluster_5': cluster_markers[5],
        'cluster_6': cluster_markers[6],
        'cluster_7': cluster_markers[7],
        'cluster_8': cluster_markers[8],
        'cluster_9': cluster_markers[9],
        'cluster_10': cluster_markers[10]}
  
df = pd.DataFrame(dict([(key, pd.Series(value)) for key, value in data.items()]))
```

```python
df
```

```python
df.to_csv('cluster_markers_signatures.csv',index=False, header=True)
```

# Simple example

```python
df = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', None, 'Indomie', 'Indomie', 'alone'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack', 'sure', None]
})
df
```

```python
com = [df['brand'],df['style']]
result = pd.concat(com)
result = result[result.duplicated()].dropna().unique()
result
```

```python
unique = pd.concat(com).unique()
unique = [item for item in unique if not(pd.isnull(item)) == True] ## retirer les valeurs None/Nan
unique
```

```python
main_list = list(set(unique) - set(result))
main_list = [item for item in (main_list) if not(pd.isnull(item)) == True]

len(main_list)
```

```python
cluster_markers = []

cluster_markers.append(list(set(df['brand']) - set(result)))
cluster_markers[0]= [item for item in cluster_markers[0] if not(pd.isnull(item)) == True]

print(str(len([item for item in df['brand'] if not(pd.isnull(item)) == True])) +" "+ str(len(cluster_markers[0])))

cluster_markers.append(list(set(df['style']) - set(result)))
cluster_markers[1]= [item for item in cluster_markers[1] if not(pd.isnull(item)) == True]

print(str(len([item for item in df['style'] if not(pd.isnull(item)) == True])) +" "+ str(len(cluster_markers[1])))


cluster_markers
```

```python
# Define a dictionary

data = {'cluster_0': cluster_markers[0],
        'cluster_1': cluster_markers[1],}
  
df = pd.DataFrame(dict([(key, pd.Series(value)) for key, value in data.items()]))
df
```
